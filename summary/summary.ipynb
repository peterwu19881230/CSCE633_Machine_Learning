## Prep data

```python
import numpy as np
import pandas as pd
accuracy_df=pd.read_csv('accuracy_df.csv',header=0,index_col=0)
accuracy_df.columns=['Pwy','pcomplex','operon','regulator_name','kegg_modules']
precision_df=pd.read_csv('precision_df.csv',header=0,index_col=0)
precision_df.columns=['Pwy','pcomplex','operon','regulator_name','kegg_modules']
```

```python
rowIndex=np.resize(['Nichols','Nichols + Price'], 12)
rowIndex_method=['Logistic Regression','Logistic Regression',
                 'Decision Tree','Decision Tree',
                 'Random Forest','Random Forest',
                 'Gradient Boosting','Gradient Boosting',
                 'SVM','SVM',
                 'CNN', 'CNN']
```

```python
#accuracy
accuracy_df.index=rowIndex
accuracy_df['rowIndex_method']=rowIndex_method
accuracy_df=accuracy_df.set_index([accuracy_df.index,'rowIndex_method'])
accuracy_df
```

```python
#precision
precision_df.index=rowIndex
precision_df['rowIndex_method']=rowIndex_method
precision_df=precision_df.set_index([precision_df.index,'rowIndex_method'])
precision_df
```

```python
import matplotlib.pyplot as plt
```

```python
df=accuracy_df
dataset='Nichols'
metric="Accuracy"


df_subset=np.array([indices[0] for indices in df.index])==dataset
df_subset=df.iloc[df_subset,:]
plt.ylim(bottom=0, top=1.1)
plt.grid(True)

x=['Pathway','protein complex','operon','regulon','kegg modules']

for i in range(df_subset.shape[0]):
    y=df_subset.iloc[i,:]
    plt.scatter(x, y,alpha=0.4)
    plt.plot(x, y,alpha=0.4)

plt.gca().legend(('Logistic Regression','Decision Tree','Random Forest','Gradient Boosting','SVM','CNN'),prop={'size': 8},framealpha=1)
plt.title(metric+' - '+dataset)
plt.ylabel(metric)

plt.savefig(metric+dataset+'.png', dpi=300)
```
